```python
import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
# load the ecg data
ecg_df = joblib.load('./ecg100')
ecg_df
```

```python
# load the patients data
patient_info = pd.read_csv('preprocess_ptbxl.csv')
patient_info
```

```python
patient_info.super_class.value_counts().plot(kind='bar')
```

```python
df = pd.merge(patient_info, ecg_df )
```

```python
df.shape
```

```python
df.query('ecg_id==21829')
```

```python
gorupby_superclass_channel = df.drop(columns=['sex', 'age_group', 'ecg_id']).groupby(['super_class', 'channels']).mean().mean(axis=1)
```

```python
plt.figure(figsize=(14,6))
plt.plot(gorupby_superclass_channel.reset_index().query('super_class=="NORM"').drop(columns='super_class').set_index('channels'), label='NORM')
plt.plot(gorupby_superclass_channel.reset_index().query('super_class=="CD"').drop(columns='super_class').set_index('channels'), label='CD')
plt.plot(gorupby_superclass_channel.reset_index().query('super_class=="MI"').drop(columns='super_class').set_index('channels'), label='MI')
plt.plot(gorupby_superclass_channel.reset_index().query('super_class=="STTC"').drop(columns='super_class').set_index('channels'), label='STTC')
plt.plot(gorupby_superclass_channel.reset_index().query('super_class=="HYP"').drop(columns='super_class').set_index('channels'), label='HYP')
plt.legend()
```

```python
# look at channel level difference
gorupby_superclass_channel = df.drop(columns=['sex', 'age_group', 'ecg_id']).groupby(['super_class', 'channels']).mean()
```

```python
gorupby_superclass_channel
```

```python
for ch in ecg_df.channels.unique():
    plt.figure(figsize=(16, 6))
    plt.plot(gorupby_superclass_channel.loc['NORM', ch], label='NORM')
    plt.plot(gorupby_superclass_channel.loc['MI', ch], label='MI')
    plt.plot(gorupby_superclass_channel.loc['STTC', ch], label='STTC')
    plt.plot(gorupby_superclass_channel.loc['CD', ch], label='CD')
    plt.plot(gorupby_superclass_channel.loc['HYP', ch], label='HYP')
    plt.title(f'channel : {ch}')
    plt.legend()
    plt.show()
```
